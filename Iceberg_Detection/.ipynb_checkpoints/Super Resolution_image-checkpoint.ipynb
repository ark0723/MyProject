```python
import json, os
from pprint import pprint

from PIL import Image

dataset = json.load(open('train.json'))

minv = -46
maxv = +35
to = 255
c = to / (maxv - minv)

min_angle = 10000
max_angle = -min_angle
na_angle_count = 0

save_dir = 'dcscn-super-resolution-master/data/band1/'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

save_dir = 'dcscn-super-resolution-master/data/band2/'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
for data in dataset:
    isIce= data[u'is_iceberg']
    b1 = data[u'band_1'] 
    b2 = data[u'band_2']
    uid = data[u'id']
    inc_angle = data[u'inc_angle']
    
    if inc_angle != u"na":
        iinc_angle = int(inc_angle)
        if min_angle > iinc_angle:
            min_angle = iinc_angle
        if max_angle < iinc_angle:
            max_angle = iinc_angle
        print(iinc_angle)
    else:
        print(inc_angle)
        na_angle_count += 1
                
    img1 = Image.new( 'L', (75,75), "black")
    img2 = Image.new( 'L', (75,75), "black")
    #img3 = Image.new( 'RGB', (75,75), "black")
    #img4 = Image.new( 'L', (75,75), "black")
    pixels1 = img1.load()
    pixels2 = img2.load()
    #pixels3 = img3.load()
    #pixels4 = img4.load()

    for i in range(0,75):
        for j in range(0,75):
            index = i * 75 + j
            v1 = int((b1[index] - minv) * c)
            v2 = int((b2[index] - minv) * c)
            pixels1[i,j] = v1
            pixels2[i,j] = v2
            #pixels3[i,j] = int((v1 + v2)/2)
            #pixels4[i,j] = v1 - v2
                
    img1.save('dcscn-super-resolution-master/data/band1/'+ uid +'.jpg')
    img2.save('dcscn-super-resolution-master/data/band2/'+ uid +'.jpg')
    #img3.save('iceberg\\'+ uid +'3.bmp')
    #img4.save('iceberg\\'+ uid +'4.bmp')
        
print(min_angle, max_angle)
print(na_angle_count, "NAs")
```
