```python
import pickle
import numpy as np
import tensorflow as tf
import email_sending
from PIL import Image, ImageDraw
import random
```

```python
#pca: unsupervised learning(X1, X2)
def band_pca():
    with open('band', 'rb') as f:
        data = pickle.load(f)
        band1 = data[0]
        band2 = data[1]

    return band1, band2
```

```python
#band1
X1 = band_pca()[0]
X2 = band_pca()[1]
```

```python
img1 = X1
img2 = X2
img1 = img1.reshape(-1,75,75)
img2 = img2.reshape(-1,75,75)
```

```python
X1 = X1.reshape(-1, 5625)
X2 = X2.reshape(-1, 5625)
```

```python
img1.shape
```

```python
from sklearn.decomposition import PCA
```

```python
#Running PCA retaining 95% of the variance 
pca_digits1=PCA(676) 
X1_proj = pca_digits1.fit_transform(X1) 
print (X1.shape, X1_proj.shape)
print (np.sum(pca_digits1.explained_variance_ratio_))
```

```python
pca_digits2=PCA(1024) 
X2_proj = pca_digits2.fit_transform(X2) 
print (X2.shape, X2_proj.shape)
print (np.sum(pca_digits2.explained_variance_ratio_))
```

#Running PCA retaining 95% of the variance 
pca_digits=PCA(0.95) 
X2_proj = pca_digits.fit_transform(X2) 
print (X2.shape, X2_proj.shape)

```python
#reconstruct the images using the new reduced dataset
#64x64 pixel images into 8x8 images. 
#Now to visualize how these images look we need to inverse transform the 8x8 images 
#back to 64x64 dimension. Note that we're not reverting back to the original data, we're simply going back to the 
#actual dimension of the original images so we can visualize them.

X1_inv_proj = pca_digits1.inverse_transform(X1_proj) 
X2_inv_proj = pca_digits2.inverse_transform(X2_proj) 
#reshaping as 400 images of 64x64 dimension 
X1_proj_img = np.reshape(X1_inv_proj,(-1,75,75)) 
X2_proj_img = np.reshape(X2_inv_proj,(-1,75,75))
```

```python
#band1 reconstruction 

fig = plt.figure(figsize=(6,6)) 
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05) 
# plot the faces, each image is 75 by 75 dimension but 26x26 pixels 
for i in range(64): 
    ax = fig.add_subplot(8, 8, i+1, xticks=[], yticks=[]) 
    ax.imshow(X1_proj_img[i], cmap=plt.cm.bone, interpolation='nearest') 
# This is not bad at all, the image still looks pretty good but the finer details are missing, which is okay considering 
# we've reduced dimensionality by 64 times. 
```
