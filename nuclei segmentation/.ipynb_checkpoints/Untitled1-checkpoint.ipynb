```python
def RGBA_To_RGB(file, save_dir):
    img = Image.open(file)
    arr = np.array(img)
    
    if img.mode =='RGB':
        img.save(save_dir,'png')
    
    if img.mode=='RGBA':
        arr= np.array(img)
        rgb = np.zeros((arr.shape[0], arr.shape[1], 3), dtype= np.uint8)
        # get row number
        for row in range(len(arr)):
            for col in range(len(arr[row])):
                for k in range(3):
                    rgb[row][col][k]= arr[row][col][0]
    
    im = Image.fromarray(rgb)
    im.save(save_dir, 'png')
```

```python
def RGBA_To_RGB(image_dir): #arr=mask array
    img = Image.open(image_dir)
    arr = np.array(img)
    
    if img.mode=='RGBA':
        arr= np.array(img)
        rgb = np.zeros((arr.shape[0], arr.shape[1], 3), dtype= np.uint8)
        # get row number
        for row in range(len(arr)):
            for col in range(len(arr[row])):
                for k in range(3):
                    rgb[row][col][k]= arr[row][col][0]
    
    im = Image.fromarray(rgb)
    return im
```
